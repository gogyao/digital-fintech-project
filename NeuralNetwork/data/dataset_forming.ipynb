```python
import pandas as pd
import glob
```

```python
files = glob.glob('*.csv')
df_all = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)
```

```python
df_all
```

```python
df_all.dropna(inplace=True)
```

```python
df_all['date'].dtype
```

```python
df_all['date'] = pd.to_datetime(df_all['date'], format='%d.%m.%Y', errors='coerce')
```

```python
df_all.sort_values(by='date', ascending=False, inplace=True)
```

```python
df_all
```

```python
df_all.to_csv('news_full.csv', index=False)
```
