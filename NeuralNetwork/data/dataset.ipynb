```python
import pandas as pd
```

```python
df = pd.read_excel("tg.xlsx")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='%d.%m.%Y', errors='coerce')
df = df.sort_values(by='Date', ascending=False)
df.drop(['USD_quant', 'RUB_quant'], axis= 1, inplace=True)
df
```

```python
df['usd-variability'] = df['USD'].pct_change() * 100
df['usd-variability'] = df['usd-variability'].round(2)

df['rub-variability'] = df['RUB'].pct_change() * 100
df['rub-variability'] = df['rub-variability'].round(2)
df
```

```python
brent = pd.read_csv("RBRTEd.csv")
brent.drop([0, 1], inplace=True)
brent
```

```python
brent.rename(columns={'Back to Contents': 'Date', 'Data 1: Europe Brent Spot Price FOB (Dollars per Barrel)': 'Brent'}, inplace=True)
brent['Date'] = pd.to_datetime(brent['Date'])
brent = brent[brent['Date'].dt.year >= 2020]
brent = brent.sort_values(by='Date', ascending=False)
brent
```

```python
df = pd.merge(df, brent, on='Date',  how='outer')
merged = df.sort_values(by='Date', ascending=False).reset_index(drop=True)
merged
```

```python
merged['Brent'] = merged['Brent'].astype('float64')
merged.loc[merged.index[0], 'Brent'] = 67.3
merged['Brent'] = merged['Brent'].interpolate(method='linear')
merged
```

```python
merged['tonia-variability'] = merged['tonia'].pct_change() * 100
merged['tonia-variability'] = merged['tonia-variability'].round(2)
merged
```

```python
merged = merged[merged['Date'] <= '2025-10-10'].reset_index(drop=True)
merged
```

```python
merged.to_csv('tenge.csv')
```
